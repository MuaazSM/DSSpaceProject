```python
# TASK 3: Calculate the distances between a launch site to its proximities
# For a single launch site
from math import radians, sin, cos, sqrt, asin

def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance between two points 
    on the earth specified in decimal degrees
    """
    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371 # Radius of earth in kilometers
    
    return c * r

def calculate_distances_to_coastline(launch_site_lat, launch_site_lon):
    """
    Calculate distances from a launch site to various coastal points
    Returns the minimum distance to coastline
    """
    # Sample coastal points
    coastal_points = {
        'East Coast': [
            (25.7617, -80.1918),  # Miami
            (32.0809, -81.0912),  # Savannah
            (39.9526, -75.1652)   # Philadelphia
        ],
        'West Coast': [
            (34.0522, -118.2437), # Los Angeles
            (37.7749, -122.4194), # San Francisco
            (47.6062, -122.3321)  # Seattle
        ]
    }
    
    min_distance = float('inf')
    closest_point = None
    
    for coast, points in coastal_points.items():
        for point in points:
            distance = calculate_distance(launch_site_lat, launch_site_lon, point[0], point[1])
            if distance < min_distance:
                min_distance = distance
                closest_point = (coast, point)
    
    return min_distance, closest_point

def calculate_distances_to_cities(launch_site_lat, launch_site_lon):
    """
    Calculate distances from a launch site to major cities
    """
    major_cities = {
        'Houston': (29.7604, -95.3698),
        'Orlando': (28.5383, -81.3792),
        'San Francisco': (37.7749, -122.4194)
    }
    
    distances = {}
    for city, coords in major_cities.items():
        distance = calculate_distance(launch_site_lat, launch_site_lon, coords[0], coords[1])
        distances[city] = distance
        
    return distances

def analyze_launch_site_proximities(launch_site_row):
    """
    Analyze proximities for a given launch site
    """
    lat = launch_site_row['Lat']
    lon = launch_site_row['Long']
    site_name = launch_site_row['Launch Site']
    coast_distance, closest_coastal_point = calculate_distances_to_coastline(lat, lon)
    city_distances = calculate_distances_to_cities(lat, lon)
    equator_distance = calculate_distance(lat, lon, 0, lon)
    
    results = {
        'Launch Site': site_name,
        'Distance to Coast (km)': round(coast_distance, 2),
        'Closest Coast': closest_coastal_point[0],
        'Distance to Equator (km)': round(equator_distance, 2),
        'Distances to Cities (km)': {city: round(dist, 2) for city, dist in city_distances.items()}
    }
    
    return results

site = launch_sites_df.iloc[0]  
results = analyze_launch_site_proximities(site)

# Print results
print(f"Analysis for {results['Launch Site']}:")
print(f"Distance to nearest coastline: {results['Distance to Coast (km)']} km")
print(f"Closest coast: {results['Closest Coast']}")
print(f"Distance to equator: {results['Distance to Equator (km)']} km")
print("\nDistances to major cities:")
for city, distance in results['Distances to Cities (km)'].items():
    print(f"  - {city}: {distance} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# Define coordinates of major cities
cities = {
    'Houston': {'coordinates': [29.7604, -95.3698], 'color': 'blue'},
    'Orlando': {'coordinates': [28.5383, -81.3792], 'color': 'green'},
    'Tampa': {'coordinates': [27.9506, -82.4572], 'color': 'purple'}
}

# For each launch site
for index, row in launch_sites_df.iterrows():
    launch_site = [row['Lat'], row['Long']]
    
    # For each city
    for city_name, city_info in cities.items():
        # Calculate distance
        distance = calculate_distance(
            launch_site[0], launch_site[1],
            city_info['coordinates'][0], city_info['coordinates'][1]
        )
        
        # Create marker at city location
        city_marker = folium.Marker(
            city_info['coordinates'],
            icon=DivIcon(
                icon_size=(20,20),
                icon_anchor=(0,0),
                html=f'<div style="font-size: 12; color:{city_info["color"]};"><b>{city_name}: {distance:.2f} KM</b></div>'
            )
        )
        
        # Create line connecting launch site to city
        line = folium.PolyLine(
            locations=[launch_site, city_info['coordinates']],
            weight=1,
            color=city_info['color'],
            opacity=0.5,
            dash_array='5, 10'  # Creates a dashed line
        )
        
        # Add both marker and line to map
        site_map.add_child(city_marker)
        site_map.add_child(line)

# Display the map
site_map
```
